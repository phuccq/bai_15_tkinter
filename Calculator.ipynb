```python
import tkinter as tk
from tkinter import ttk as ttk


def press(num):
    global expression
    expression+=num
    equation.set(expression)
    
def calculate():
    try:
        total = eval(expression)
        equation.set(total)
    except:
        equation.set('error')
        
def clear():
    global expression
    expression = ''
    equation.set(expression)
        
root=tk.Tk()
root.geometry('350x250')
root.title('Calculator')
root.resizable(False,False)


expression=''
equation=tk.StringVar()
equation.set('')


frame1=ttk.Frame(root)
entry=ttk.Entry(frame1,textvariable=equation)
entry.pack(side='left',fill='both',expand=True, padx='5',pady='10')

frame2=ttk.Frame(root)
button_1=ttk.Button(frame2,text='1',command=lambda:press('1'))
button_2=ttk.Button(frame2,text='2',command=lambda:press('2'))
button_3=ttk.Button(frame2,text='3',command=lambda:press('3'))
button_plus=ttk.Button(frame2,text='+', command=lambda:press('+'))
button_1.pack(side='left',fill='both', padx='5',pady='5')
button_2.pack(side='left',fill='both', padx='5',pady='5')
button_3.pack(side='left',fill='both', padx='5',pady='5')
button_plus.pack(side='left',fill='both', padx='5',pady='5')

frame3=ttk.Frame(root)
button_4=ttk.Button(frame3,text='4',command=lambda:press('4'))
button_5=ttk.Button(frame3,text='5',command=lambda:press('5'))
button_6=ttk.Button(frame3,text='6',command=lambda:press('6'))
button_minus=ttk.Button(frame3,text='-', command=lambda:press('-'))
button_4.pack(side='left',fill='both', padx='5',pady='5')
button_5.pack(side='left',fill='both', padx='5',pady='5')
button_6.pack(side='left',fill='both', padx='5',pady='5')
button_minus.pack(side='left',fill='both', padx='5',pady='5')

frame4=ttk.Frame(root)
button_7=ttk.Button(frame4,text='7',command=lambda:press('7'))
button_8=ttk.Button(frame4,text='8',command=lambda:press('8'))
button_9=ttk.Button(frame4,text='9',command=lambda:press('9'))
button_multiply=ttk.Button(frame4,text='*',command=lambda:press('*'))
button_7.pack(side='left',fill='both', padx='5',pady='5')
button_8.pack(side='left',fill='both', padx='5',pady='5')
button_9.pack(side='left',fill='both', padx='5',pady='5')
button_multiply.pack(side='left',fill='both', padx='5',pady='5')

frame5=ttk.Frame(root)
button_0=ttk.Button(frame5,text='0',command=lambda:press('0'))
button_clear=ttk.Button(frame5,text='Clear', command=clear)
button_equal=ttk.Button(frame5,text='=', command=calculate)
button_devide=ttk.Button(frame5,text='/', command=lambda:press('/'))
button_0.pack(side='left',fill='both', padx='5',pady='5')
button_clear.pack(side='left',fill='both', padx='5',pady='5')
button_equal.pack(side='left',fill='both', padx='5',pady='5')
button_devide.pack(side='left',fill='both', padx='5',pady='5')

frame6 =ttk.Frame(root)
button_dot=ttk.Button(frame6, text='.',command=lambda:press('.'))
button_dot.pack(side='left',fill='both', padx='5',pady='5')

frame1.pack(side='top',fill='both', padx='10')
frame2.pack(side='top',fill='both', padx='10')
frame3.pack(side='top',fill='both', padx='10')
frame4.pack(side='top',fill='both', padx='10')
frame5.pack(side='top',fill='both', padx='10')
frame6.pack(side='top',fill='both', padx='10')

tk.mainloop()
```
