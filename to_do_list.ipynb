```python
import tkinter as tk
from tkinter import ttk as ttk

main = tk.Tk()
main.title('To do list')
main.geometry('300x200')
main.resizable(False, False)

items = []
entry_value = tk.StringVar()
list_items = tk.StringVar()

def add_item():
    global items
    items.append(entry_value.get())
    list_items.set(items)
    entry_value.set('')
    
def delete_item():
    global items
    delete_index = to_do_list.curselection()
    count=0
    for index in delete_index:
        if count == 0:
            items.pop(index)
        else:
            items.pop(index-count)
        count+=1        
    list_items.set(items)        
    
def delete_all_item():
    global items
    items.clear()
    list_items.set(items)

def exit():
    main.destroy()

frame1 = ttk.Frame(main)
title_text = ttk.Label(frame1, text='To-Do List')
entry_text = ttk.Label(frame1, text='Enter task title')
entry = ttk.Entry(frame1, textvariable = entry_value)
add_button = ttk.Button(frame1, text = 'Add task', command = add_item)
del_button = ttk.Button(frame1, text = 'Delete', command = delete_item)
del_all_button = ttk.Button(frame1, text = 'Delete all', command = delete_all_item)
exit_button = ttk.Button(frame1, text = 'Exit', command = exit)

title_text.pack(side = 'top', fill = 'both')
entry_text.pack(side = 'top', fill = 'both')
entry.pack(side = 'top', fill = 'both')
add_button.pack(side = 'top', fill = 'both')
del_button.pack(side = 'top', fill = 'both')
del_all_button.pack(side = 'top', fill = 'both')
exit_button.pack(side = 'top', fill = 'both')

frame2 = ttk.Frame(main)
blank_text = ttk.Label(frame2, text='')
to_do_list = tk.Listbox(frame2, listvariable = list_items, selectmode = 'extended', height=150)
blank_text.grid(row=0,column=0)
to_do_list.grid(row=1,column=0)

frame1.pack(side = 'left', fill = 'both', padx=10, pady=10)
frame2.pack(side = 'left', fill = 'both', padx=10, pady=10)

tk.mainloop()
```
